```python
causes = pd.read_csv('C:\\Users\\19713\\Desktop\\Springboard\\Causes of maternal death, 2014.csv')
```

```python
causes.rename(columns = {'Unnamed: 0' : 'Cause of Maternal Death', 'Unnamed: 1' : 'Proportion (in percentages)'}, inplace = True)
```

```python
causes
```

```python
causes.groupby(['Cause of Maternal Death']).sum().plot(kind='pie', y='Proportion (in percentages)', autopct='%1.0f%%',
                                colors = ['red', 'pink', 'steelblue'],
                                title='Causes of Maternal Death')
```

This pie chart shows the causes of maternal death along the proportion in percentages of each cause. This allows us to compare the different causes according to their prevelancy and divert our attention and focus on the more common causes of maternal death. 

```python
lifetime_risk_region = pd.read_csv('C:\\Users\\19713\\Desktop\\Springboard\\Lifetime risk of maternal death_ 1 in x, region_group.csv')
```

```python
lifetime_risk_region.rename(columns = {'Unnamed: 0' : 'Region', '2020': 'Lifetime risk of maternal death: 1 in X'}, inplace = True)
```

```python
lifetime_risk_region
```

```python
lifetime_risk_region.plot.barh(x = 'Region', y = 'Lifetime risk of maternal death: 1 in X')
```

This bar chart shows how likely a mother is to die giving birth depending on in which region they live in. The lifetime risk of maternal death here is expressed as the probability that a 15-year-old girl will die from complications of preganancy or childbirth over her lifetime. The smaller the bar is in this chart, the more likely it is that the mother is to die from complications of preganancy or childbirth. We can see from the chart that women from West and Central Africa are the most likely to die from maternal death; on the other end of the spectrum, a mother from Western Europe is the least likely to perish from maternal death.
